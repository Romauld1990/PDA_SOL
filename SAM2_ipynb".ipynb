11) Создать матрицу случайных чисел размера 10 на 10 и отнять медиану по столбцам из каждого элемента столбца  

```python
import numpy as np

# создаем матрицу случайных чисел размера 10 на 10
matrix = np.random.rand(10, 10)

# вычисляем медиану элементов каждого столбца
medians = np.median(matrix, axis=0)

# отнимаем медиану из каждого элемента столбца
result = matrix - medians

print(result)
```

12) Создать матрицу случайных чисел 5x5. Добавить к ней столбецы максимумов и минимумов по строкам. Получится матрица 5x7.  

```python
# Создание матрицы случайных чисел размера 10 на 10
matrix = np.random.randint(1, 10, (10, 10))
print("Исходная матрица:\n", matrix)

# Вычисление среднего значения для каждой строки
mean = np.mean(matrix, axis=1)

# Вычитание среднего значения из каждого элемента строки
matrix = matrix - mean.reshape(-1, 1)

print("Матрица после вычитания среднего значения по строкам:\n", matrix)
```

13) Вектор чисел от 1 до 51 преобразовать в вектор нечетных чисел и в вектор четных чисел, поменять их размерности, чтобы получились две матрица. Найти произведение этих матриц. Найти обратную матрицу эётого произведения.  

```python
import numpy as np

# Создаем вектор чисел от 1 до 51
v = np.arange(1, 52)

import numpy as np

# Создаем вектор чисел от 1 до 51
v = np.arange(1, 52)

# Создаем вектор нечетных чисел
odd_v = v[v%2 != 0]

# Создаем вектор четных чисел
even_v = v[v%2 == 0]

# Меняем размерности векторов
odd_m = odd_v.reshape((len(odd_v), 1))
even_m = even_v.reshape((len(even_v), 1))

# Находим произведение матриц
prod = np.dot(odd_m, even_m.T)

print(prod)
```

14) Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.

```python
# создаем матрицу размерности 8x8
matrix = np.zeros((8, 8))

# заполняем диагональные элементы единицами
np.fill_diagonal(matrix, 1)

# заполняем элементы на четных строках и столбцах единицами
for i in range(0, 8, 2):
    for j in range(0, 8, 2):
        matrix[i][j+1] = 1
        matrix[i+1][j] = 1

print(matrix)
```
